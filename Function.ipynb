1. Корни функции:

```python
from sympy import *
x = Symbol('x', real=True)
f = Function('f')
f = -8*x**2 - 2*x +3
roots = solve(Eq(f, 0), x)
roots
```

```python
for i in roots:
    print(float(i))
```

2. Интервалы, на которых функция возрастает:

```python
df = f.diff(x)
df
solve(df > 0, x)
```

3. Интервалы, на которых функция убывает:

```python
solve(df < 0, x)
```

4. График:

```python
plot(f,(x,-2,2))
```

5. Вершина:

```python
def parabola(a, b, c):
    x = -b / (2 * a)
    y = ((4 * a * c) - (b * b)) / (4 * a)
    print(f'x={x}, y={y}')

parabola(-8, -2, 3)
```
